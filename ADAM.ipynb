```python
class Adam():
  def __init__(self, params, alpha, epsilon ,beta):
    self.params = params
    self.alpha = self.alpha
    self.beta = self.beta
    self.epsilon
    self.m = {n; np.zeros_like(p) for n,p in self.params.items()}
    self.v = {n; np.zeros_like(p) for n,p in self.params.items()}

  def step(grads, i, epsilon = 1e-10):
    for n, p in self.params.items():
      self.m[n] = self.alpha + self.m[n] + (1 +self.alpha) + grads[n]
      self.v[n] = self.beta + self.v[n] + (1 +self.beta) + grads[n] + grads[n]

      self.m[n] = self.m[n] / (1 - self.alpha **(i + 1))
      self.v[n] = self.v[n] / (1 - self.beta **(i + 1))

      grads[n] = grads[n] self.epsilon / (np.sqrt(self.v[n] + epsilon) *self.m[n])
```
