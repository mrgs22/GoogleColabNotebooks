```python
#preprocessing
#grabbed most preprocessing methods from here || https://spotintelligence.com/2022/12/21/nltk-preprocessing-pipeline/
stopwords = nltk.corpus.stopwords.words("english")
lemmatizer = nltk.stem.WordNetLemmatizer()
exclude_words = ["I", "he", "she", "it", "we", "they", "a", "an", "the", "''", "``"] #this is need for line 19
def preprocessing (text, lemmatization=True):
  pattern = r"(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?"
  cleaned_text = re.sub(pattern, "", text)
  pattern = r"<[^>]+>"
  cleaned_text = re.sub(pattern, "", text)
  text = text.strip()
  text = " ".join(text.split())
  tokens = nltk.word_tokenize(text)
  tokens = [token.lower() for token in tokens]
  tokens = [token.strip(string.punctuation) for token in tokens]
  tokens = [token for token in tokens if token not in string.punctuation]
  tokens = [token for token in tokens if token.lower() not in stopwords]
  if lemmatization == True:
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
  tokens = [token for token in tokens if (len(token) > 1 or token.lower() in exclude_words) or re.match(r'^\d+$', token)] # this line I got help from chatgpt it helps to remove chorter tokens that don't make sense or are not words/numbers
  return tokens
```

```python
test_tokens = []
vocab_counter = Counter()
#for i in tqdm(range(len(dataset["train"]))):
for i in tqdm(range(len(dataset["train"]))):
  tokens = preprocessing(dataset["train"][i].get('text'))
  test_tokens.append(tokens)
  vocab_counter.update(tokens)
```

```python
#models below
model_CBOW = Word2Vec(sentences=test_tokens, vector_size=vector_size, window=window, min_count=min_count,workers= workers)
model_SG = Word2Vec(sentences=test_tokens, sg = 1, vector_size=vector_size, window=window, min_count=min_count,workers= workers)
```

```python
#section 2.4
#Bias in Word Embeddings
query = Query(
    target_sets = [
        ["woman", "female", "girl", 'gale'],
        ["man", 'male', 'boy', 'guy']
    ],
    attribute_sets=[
        ["smart", "sharp", "bright", "clever"],
        ["strong", "powerful", "tough", "fit"]
    ],
    target_sets_names=["female terms", "male terms"],
    attribute_sets_names= ["brain", "strength"]
)
```

```python
#this function is from the example use in the github || https://github.com/dccuchile/wefe/blob/develop/examples/RNSB_experiments.ipynb
def evaluate(query: Query, modelname):
    model = WordEmbeddingModel(modelname)

    # Run the queries
    results = RNSB().run_query(
        query, model, holdout=True, print_model_evaluation=True, n_iterations=1000
    )

    # Show the results obtained with glove
    fig = px.bar(
        pd.DataFrame(
            results["negative_sentiment_distribution"].items(),
            columns=["Word", "Sentiment distribution"],
        ),
        x="Word",
        y="Sentiment distribution",
        title=f"{modelname} Negative Sentiment Distribution",
    )

    fig.update_yaxes(range=[0, 0.2])
    fig.show()
```

```python
evaluate(query, CBOW_wv)
evaluate(query, sg_wv)
evaluate(query, gw_wv)
evaluate(query, gt_wv)
```

```python
def cbow_vectorizers(texts, wv):
    vectors = []
    for text in tqdm(texts, desc="Vectorizing text"):
        tokens = preprocessing(text)
        doc_vectors = []
        for token in tokens:
            if token in wv:
                doc_vectors.append(wv[token])
        if doc_vectors:
            doc_mean_vector = np.mean(doc_vectors, axis=0)
            vectors.append(doc_mean_vector)
        else:
            vectors.append(np.zeros(wv.vector_size))
    return np.array(vectors)
```
