```python
#Imports

import numpy as np
import requests
from bs4 import BeautifulSoup
import re
import spacy
```

```python
#import existing text lines of charcter from game
url = "https://theportalwiki.com/wiki/Wheatley_voice_lines"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    voice_lines = soup.find_all("li")
    text = []
    for line in voice_lines:
        # Use regular expression to find text within quotation marks
        phrases = re.findall(r'"(.*?)"', line.text.strip())

        # Append the extracted quoted phrases to the array
        text.extend(phrases)

    # Print the extracted quoted text
    for phrase in text:
        print(phrase)

    training_text = np.array(text)

else:
    print("Failed to retrieve webpage. Status code:", response.status_code)
```

```python
capitalized_pattern = re.compile(r'\b[A-Z][A-Z]+\b')
nlp = spacy.load("en_core_web_sm")
def preprocess(text_array):
    preprocessed_texts = []
    for text in text_array:
        doc = nlp(str(text))

        preprocessed_tokens = []

        for token in doc:
            if re.match(capitalized_pattern, token.text):
                preprocessed_tokens.append(token.text)
            else:
                preprocessed_tokens.append(token.text.lower())

        preprocessed_text = ' '.join(preprocessed_tokens)
        preprocessed_texts.append(preprocessed_text)

    return np.array(preprocessed_texts)

preprocessed_training_text = preprocess(training_text)

print(preprocessed_training_text)
```

```python
def remove_duplicates(corpus):
    # Initialize a set to store unique document texts
    unique_texts = set()

    # Initialize a list to store non-duplicate documents
    deduplicated_corpus = []

    # Iterate through each document in the corpus
    for document in corpus:
        # Check if the document text is not in the set of unique texts
        if document not in unique_texts:
            # Add the document text to the set of unique texts
            unique_texts.add(document)
            # Add the document to the deduplicated corpus
            deduplicated_corpus.append(document)

    return deduplicated_corpus
```

```python
corpus = remove_duplicates(preprocessed_training_text)
```

```python
print(corpus)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import matplotlib.pyplot as plt

# Sample dataset of Wheatley-like dialogue
dialogue = training_text

# Tokenize dialogue
tokenizer = Tokenizer()
tokenizer.fit_on_texts(dialogue)
total_words = len(tokenizer.word_index) + 1

# Create input sequences and labels
input_sequences = []
for line in dialogue:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

# Create predictors and label
xs, labels = input_sequences[:,:-1], input_sequences[:,-1]
ys = tf.keras.utils.to_categorical(labels, num_classes=total_words)

# Define model architecture
model = tf.keras.Sequential([
    Embedding(total_words, 100, input_length=max_sequence_len-1),
    LSTM(150),
    Dense(total_words, activation='softmax')
])

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
history = model.fit(xs, ys, epochs=40, verbose=1)

# Plot training accuracy
plt.plot(history.history['accuracy'])
plt.title('Model Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dense, Embedding, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset of Wheatley-like dialogue
dialogue = training_text

# Tokenize dialogue
tokenizer = Tokenizer()
tokenizer.fit_on_texts(dialogue)
total_words = len(tokenizer.word_index) + 1

# Create input sequences and labels
input_sequences = []
for line in dialogue:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i+1]
        input_sequences.append(n_gram_sequence)

# Pad sequences
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))

# Create predictors and label
xs, labels = input_sequences[:,:-1], input_sequences[:,-1]
ys = tf.keras.utils.to_categorical(labels, num_classes=total_words)

# Define model architecture
model = tf.keras.Sequential([
    Embedding(total_words, 100, input_length=max_sequence_len-1),
    Bidirectional(LSTM(150)),
    Dense(total_words, activation='softmax')
])

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
history = model.fit(xs, ys, epochs=4, verbose=1)

# Plot training accuracy
plt.plot(history.history['accuracy'])
plt.title('Model Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.show()
```
